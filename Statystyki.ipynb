```python
!pip install scikit-learn
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
# Generowanie danych z rozkładu normalnego
mu = 3
sigma = 1
# Próbka 100 wartości
data_100 = np.random.normal(mu, sigma, 100)
# Próbka 10000 wartości
data_10000 = np.random.normal(mu, sigma, 10000)
```

```python
sns.distplot(data_100)
```

```python
sns.distplot(data_10000)
```

```python
mean_data_100 = np.round(np.mean(data_100))
print(mean_data_100)
```

```python
mean_data_10000 = np.round(np.mean(data_10000))
print(mean_data_10000)
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x = mean_data_100, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x = mean_data_10000, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
print("Średnie obu zbiorów są takie same, co wynika bezpośrednio z założenia samego zadania, że zbiory miały mieć średnią '3'.")
```

```python
median_data_100 = np.median(data_100)
print(median_data_100)
```

```python
median_data_10000 = np.median(data_10000)
print(median_data_10000)
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x = mean_data_100, color='red', label = 'Średnia')
plt.axvline(x = median_data_100, color='green', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x = mean_data_10000, color='red', label = 'Średnia')
plt.axvline(x = median_data_10000, color='green', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
print("Mediany obu zbiorów różnią się nieznacznie, co zapewne wynika z liczebności próby.")
```

```python
mode_data_100 = stats.mode(data_100)
print(mode_data_100)
```

```python
mode_data_10000 = stats.mode(data_10000)
print(mode_data_10000)
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x = mean_data_100, color='red', label = 'Średnia')
plt.axvline(x = median_data_100, color='green', label = 'Mediana')
plt.axvline(x = mode_data_100[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x = mean_data_10000, color='red', label = 'Średnia')
plt.axvline(x = median_data_10000, color='green', label = 'Mediana')
plt.axvline(x = mode_data_10000[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()
```

```python
print("W obu zbiorach jako moda został potraktowany pierwszy element próby. Wygenerowane zbiory mają wszystkie wartości inne od siebie zatem najczęściej występującym elementem próby będzie wartość występująca jednokrotnie i w tym przypadku odczytana w obu zbiorach jako pierwsza wartość w próbie.")
```

```python
q0_100 = np.quantile(data_100, 0.0)
q1_100 = np.round(np.quantile(data_100, 0.25))
q3_100 = np.round(np.quantile(data_100, 0.75))
q4_100 = np.round(np.quantile(data_100, 1.0))
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_data_100}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
```

```python
q0_10000 = np.quantile(data_10000, 0.0)
q1_10000 = np.round(np.quantile(data_10000, 0.25))
q3_10000 = np.round(np.quantile(data_10000, 0.75))
q4_10000 = np.round(np.quantile(data_10000, 1.0))
print(f'Q0: {q0_10000}')
print(f'Q1: {q1_10000}')
print(f'Q2: {median_data_10000}')
print(f'Q3: {q3_10000}')
print(f'Q4: {q4_10000}')
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x = mean_data_100, color='red', label = 'Średnia')
plt.axvline(x = median_data_100, color='green', label = 'Mediana')
plt.axvline(x = mode_data_100[0], color='orange', label = 'Moda')
plt.axvline(x = q0_100, color='black', label = 'Min')
plt.axvline(x = q1_100, color='black', label = 'Q1')
plt.axvline(x = q3_100, color='black', label = 'Q3')
plt.axvline(x = q4_100, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x = mean_data_10000, color='red', label = 'Średnia')
plt.axvline(x = median_data_10000, color='green', label = 'Mediana')
plt.axvline(x = mode_data_10000[0], color='orange', label = 'Moda')
plt.axvline(x = q0_10000, color='black', label = 'Min')
plt.axvline(x = q1_10000, color='black', label = 'Q1')
plt.axvline(x = q3_10000, color='black', label = 'Q3')
plt.axvline(x = q4_10000, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
print("Wartośći kwartyli w obu zbiorach są różne dla Q0 i Q4, natomiast Q1 i Q3 mają takie same wartości, co wynika z zastosowanego rozkładu.")
```

```python
# zakresy
range_data_100 = max(data_100) - min(data_100)
print(range_data_100)
```

```python
range_data_10000 = max(data_10000) - min(data_10000)
print(range_data_10000)
```

```python
print("Rozpiętości obu prób znacznie się różnią pomimo tego, że obie próby wykazują rozkład normalny. Wynika to z tego skład elementów poszczególnych prób jest całkowicie różny.")
```

```python
# roztęp międzykwartylowy
iqr_100 = q3_100 - q1_100
print(iqr_100)
```

```python
iqr_10000 = q3_10000 - q1_10000
print(iqr_10000)
```

```python
print("W obu próbach rozkład międzykwartylowy jest taki sam co wiąże się z tym, że obie próby odznaczają się taką samą średnią i odchyleniem standardowym.")
```

```python
# wariancja
variance_data_100 = np.var(data_100, ddof=1)
print(variance_data_100)
```

```python
variance_data_10000 = np.var(data_10000, ddof=1)
print(variance_data_10000)
```

```python
print("Różne wartości wariancji wynikają z faktu, że zbiory są róźne, posiadają różne elementy.")
```

```python
# odchylenie standardowe
standard_deviation_data_100 = np.std(data_100, ddof=1)
print(standard_deviation_data_100)
```

```python
standard_deviation_data_10000 = np.std(data_10000, ddof=1)
print(standard_deviation_data_10000)
```

```python
print("Odchylenia standardowe różnia się w obu zbiorach, co jest warunkowane różnymi elementami zbiorów, pomimo tego, że oba zbiory posiadają taki sam rozkład.")
```
